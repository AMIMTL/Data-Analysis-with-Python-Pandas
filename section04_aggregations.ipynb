# Assignment 4: The Agg Method

Calculate the mean of target met by store, and the sum of bonuses to be paid to each store.

Sort them by highest to lowest bonus payout.

Then, do the same for day of week and month.

```python
# Recreate table from section 3

transaction = transaction.assign(
    target_pct=(transaction["transactions"] / 2500).round(2),
    met_target=(transaction["transactions"] / 2500) >= 1,
    bonus_payable=((transaction["transactions"] / 2500) >= 1) * 100,
    month=transaction.date.dt.month,
    day_of_week=transaction.date.dt.dayofweek,
)

transaction.head()
```

```python
transaction.groupby(["store_nbr"])[["met_target"]].mean().round(2).head()
```

```python
(transaction.groupby("store_nbr")
 .agg({
     "met_target" : "mean" , 
     "bonus_payable" : "sum"
 })
 .round(2)
 .sort_values(by = ["bonus_payable"] , ascending = 0)
).head(10)
```

```python
(transaction.groupby(["month" , "day_of_week"])
 .agg({
     "met_target" : "mean" , 
     "bonus_payable" : "sum"
 })
 .round(2)
 .sort_values(by = ["bonus_payable"] , ascending = [0])
).head(14)
```

```python
(transaction.groupby(["month" , "day_of_week"])
 .agg({
     "target_pct" : "mean" , 
     "bonus_payable" : "sum"
 })
 .round(2)
 .sort_values(by = ["month" , "day_of_week" , "bonus_payable"] , ascending = [1,1,0])
).head(14)
```

```python
(transaction.groupby(["month" , "day_of_week"])
 .agg({
     "met_target" : "mean" , 
     "bonus_payable" : "sum"})
 .round(2)
 .reset_index()
 .sort_values(by = ["bonus_payable"] , ascending = [0])
).head(14)
```

```python
import nltk
from nltk.corpus import wordnet
import seaborn as sns
import pandas as pd
import numpy as np

nltk.download('wordnet')

# Let's say your word is 'peculiar'
synonyms = []
for syn in wordnet.synsets('peculiar'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

# Create a dataframe to hold the likelihoods
df = pd.DataFrame(data=np.random.rand(len(synonyms), len(synonyms)), 
                  index=synonyms, columns=synonyms)

# Plot a heatmap
sns.heatmap(df)
```

```python
import nltk
from nltk.corpus import wordnet
import matplotlib.pyplot as plt

nltk.download('wordnet')

# Let's say your word is 'peculiar'
synonyms = []
for syn in wordnet.synsets('peculiar'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

# For simplicity, let's assign random relevance scores (in real scenario, you might use a more sophisticated method)
relevance = np.random.rand(len(synonyms))

# Create a color map
cmap = plt.get_cmap("cool")

# Create a figure and axis
fig, ax = plt.subplots()

# Create a color bar
sm = plt.cm.ScalarMappable(cmap=cmap)
plt.colorbar(sm)

# Plot each synonym with a color that corresponds to its relevance
for i, synonym in enumerate(synonyms):
    ax.text(0.5, i, synonym, color=cmap(relevance[i]))

plt.show()
```
